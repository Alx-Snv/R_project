From the dataset it can be observed that it is a campus recruitment dataset with columns displaying student's percentages obtained from 10 std. up till the area of specialization they have selected. The areas of interest whether it may be art, commerce, science or MBA & HR or MBA & FIN. Here we are able to see the structure of the dataset with each column's datatype displayed along with total number of rows & columns. Here we can observe that the columns are divided amongst three datatypes which are int, character and number. Here chr columns are called categorical columns.

```r
summary(recruitment) # Give statistical information about the dataset
```

Summary is used to give statistical information about each of the feature/columns such as min, Max, mean, median, mode, the inter-quartile range for the numerical columns and the length of observations and class type for the categorical columns.
```r
recruitment[is.na(recruitment)] = 0 # replacing NA with 0
```

The reason being that these null values represent the students who have not been placed and therefore will not have received any salary. If this was to be replaced by the mean or any other value then this will affect the analysis being performed on the dataset.

# Data Visualization
Here we have created different visualisations to gather insights from the data to observe the trend or pattern being followed during the recruitment process. We have made use of bar plots, histograms, pie and scatter plots.

```r
bar1 <- ggplot(recruitment, aes(ssc_b)) # creating a variable to store the co-ordinate to be plotted
bar1 + geom_bar(aes(fill = ssc_b)) # color the bar chart
```

Here we have made use of ggplot library as it allows use to create more customisable charts with better overall aesthetics. The first plot is a simple bar plot where we create aesthetic mapping with the ssc board(i.e. we map our selected variable ssc_b to visual properties of ggplot geoms). After creating the base, we then fill it with status to create a plot where we can see the how many students got placed and their boards centres along with the unplaced students. Here we can observe that central board students have a higher chance to get recruited against the students from other board centres

```r
bar2 <- ggplot(recruitment, aes(hsc_s))
bar2 +geom_bar(aes(fill = ssc_b)) # color the bar chart with respect to a different column to show distribution of data
```

Here we used HSC board for aes mapping along with the ssc board as fill option to observe the selection of domain by the students after clearing their ssc board exams. It can noted that the students are distributed equally among each individual domain (where 50% come from central and the remaining are from other board). But overall graph shows that students are more likely to select commerce as their preference followed closely by science and only a small minority choosing art.

```r
bar3 <- ggplot(recruitment, aes(degree_t))
bar3 +geom_bar(aes(fill = degree_t)) + geom_text(stat = "count", aes(label=after_stat(count))) # using text function to count the quantity
```

Here we have made use of count function to count the exact number of students in the degree phase. With 146 being in Commerce and management, 59 in science and technology, only 11 in other. In the previous graph it can be observed that close to 110 students had been in commerce and around 90 in science, the arts had roughly around 15 students. But now it can be inferred from the plots that some of science students may have shifted to the commerce section as there seems to be an increase in the commerce department while the science department has lost some of its numbers with the arts remaining almost the same.

```r
bar3 <- ggplot(recruitment, aes(degree_t))
bar3 +geom_bar(aes(fill = hsc_s), position = "dodge") # shifting position so that the bar is broken down according to the column given and presented on the x-axis
```

To further delve into the migration of the students we created a bar plot with the degree_t and used fill with the HSC_b. Here the position of the graph is made using dodge to get side by side plots rather than stacked bars. But moving on, we can see that the commerce students did not change their domain with the exception of few, science on the other hand did have a considerable amount of students shifting to commerce department. The art department remained the same more or less

```r
bar4 <- ggplot(recruitment, aes(specialisation))
bar4 +geom_bar(aes(fill = status), position = "dodge")
```

The specialization degree is a notable feature as it shows that people who chose MKT & FIN have received more placement offers than the other department. On the other hand MKT & HR has a low acceptance in getting placed

```r
hist1 <- ggplot(recruitment, aes(salary))
hist1 + geom_histogram(binwidth = 50000, fill = "green") # adjusting the binwidth to 50000 to better accomodate the data and then coloring the histogram in green
```

After bar plots, we have moved to the histograms to analyse the numerical columns. Here we have histogram of the salary column, with the bar on 0 depicting unplaced students. The salary packages most of the students receive according to this data is around 250000 with only a handful receiving packages around 100000. This is the norm because the higher the package is the less the amount of candidates is selected.

```r
hist1 + geom_histogram(aes(fill = status), binwidth = 50000) # instead of color using the status column to show distribution accordingly
```

```r
# using facet_wrap to create create a multi-panel plot with one pane per alignment
hist1 + geom_histogram(aes(fill = status), binwidth = 100000) + facet_wrap(~gender) # here gender is used for making a multi-panel plot with different panes for the male and female gender with the same alignment
```

From the multi-plot following can be concluded: 1. No of males getting employed is more than the females
2. The base salary for majority of the males is >250000, but for females it is<250000.
3. The highest salary packages females received were around 650000 and for males it is 100000

```r
hist1 + geom_histogram(aes(fill = status), binwidth = 100000) + facet_wrap(~workex)
```

Work experience proves to be valuable when getting placed as we can see that people with prior work experience have less chances of rejection and are also likely to receive higher salary than that of the freshers. This is due to the experience and knowledge they have gained from their previous employment making them more flexible to the work environment than the freshers who 0 experience

```r
box1 <- ggplot(recruitment, aes(y = mba_p))
box1 + geom_boxplot(fill = "yellow") # creating boxplot to show mba %
```

Creating a box plot with the MBA percentage to check the percentage distribution with the minimum being around 58 and the max around 66. There are no outliers present in this data column

```r
box1 + geom_boxplot(aes(fill = status)) # adding status to show boxplot depicting how students are placed with regards to their mba %
```

With status we can gain further knowledge such as that there is not much difference in the median of placed and unplaced students. The quartile range for placed students is more than the unplaced students.

```r
box1 + geom_boxplot(aes(fill = status)) +
  facet_wrap(~specialisation) # using the above plot but adding specialisation to create a multi-panel plot
```

MBA specialization used as facet wrap gives us a multi plot with the placement status. The Finance students with higher percentage are more likely to be placed than the unplaced ones. On the other hand it can be seen that in HR students with lower percentage seem to get placed than some of the students with higher percentage.

```r
box1 + geom_boxplot(aes(fill = status)) +
  facet_wrap(specialisation~workex) # adding 2 columns in the facet_wrap function to further look into the dataset
```

For the same box plot but with work experience as facet wrap, the same thing can be observed that in FIN people with higher percentage seem to get placed amongst their peers whereas almost the opposite can be seen in the HR

```r
scatter1 <- ggplot(recruitment, aes(x= degree_p, y = mba_p)) # create scatter plot for degree % and mba %
scatter1 + geom_point()
```

Here we are creating a scatter plot with the aes mapping the variable of degree percentage and MBA percentage.

```r
scatter1 + geom_point(aes(color = status)) # adding status to filter how many students are placed or not
```

Status column is added in the fill to create a visually colourful plot and also to see the placement status. It can be assumed that overall placed points are further down the x-axis and also the y-axis to some extent

```r
scatter1 <- ggplot(recruitment, aes(x= degree_p, y = mba_p))
scatter1 + geom_point() + geom_density_2d() # to display where the points are concentrated and how they spread out
```

The density line is drawn to show the distribution pattern of the data. It can be seen that the data points seem to be somewhat concentrated in a certain area of the plot

```r
scatter1 + geom_point() +  geom_bin2d() # to display the co-ordinate concentration
```

Bin2d is the same as density lines but in colour format to show the density of data distribution with the darker the colour the less the density and vice versa

```r
scatter1 <- ggplot(recruitment, aes(x= degree_p, y = mba_p))
scatter1 + geom_point(alpha = 0.7, color = "orange", size = 5) + geom_density_2d(color = "purple") # customizing the plot by changing point colours and the density colour
```

```r
scatter1 <- ggplot(recruitment, aes(x= degree_p, y = mba_p))
scatter1 + geom_point() + geom_smooth() # adding smooth to display trend lines over existing plot
```

Smoother is implemented to display the trend line over the data distribution. Here we can observe the trend is positive in nature with the shaded region being area of confidence. This smoother is linear in nature being that for every change in the input there will be proportional change in the output

```r
scatter1 + geom_point(aes(color = status)) + geom_smooth(aes(colour = status)) # adding trend lines w.r.t the status column
```

The two smooth lines display the placement status with the unplaced students showing a downward trend further along the x axis. But for the placed it is showing upward trend further along with the X axis

```r
scatter1 + geom_point() + geom_smooth(aes(colour = status))
```

```r
scatter1 + geom_point() + geom_smooth(aes(colour = status)) + facet_wrap(~gender) # using gender to create multi-panel plots
```

For the gender we can observe that the males are seeing upwards trends in the placed category but the females on the other hand see a slightly downward trend in the placed status. The upwards trends here in general suggest that the higher the percentage, the more likely students get placed. On the other hand downward suggest that percentage do not contribute to the selection process in job placements

```r
scatter1 + geom_point() + geom_smooth(aes(colour = status)) + facet_wrap(~workex)
```

For the work experience feature, same can be observed in the graph as it did for gender where people with prior work experience having good performance show upwards trends. Same goes for the students with 0 work experience

```r
scatter1 + geom_point() + geom_smooth(aes(colour = status)) + facet_wrap(~specialisation)
```

This is almost same as the above graph in trend line

```r
scatter1 + geom_point() + geom_smooth(aes(colour = status)) + facet_wrap(ssc_b~gender)
```
